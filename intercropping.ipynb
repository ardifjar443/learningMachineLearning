```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
encodings = ['utf-8', 'latin1', 'iso-8859-1', 'utf-16']

for encoding in encodings:
    try:
        df = pd.read_csv("./dataset/intercropping/data1.csv", encoding=encoding)
        # Jika pembacaan berhasil, keluar dari loop
        break
    except UnicodeDecodeError:
        print(f"Gagal membaca dengan encoding {encoding}. Mencoba encoding berikutnya...")
```
